```python
district_leveltotal=District_TxnandUsers.groupby('State').agg(total_transactions=('Transactions','sum'),total_amount=('Amount (INR)','sum'),total_users=('Registered Users','sum')).reset_index()
district_leveltotal
```

```python
merged_data = pd.merge(state_leveltotal, district_leveltotal, on='State', suffixes=('_state', '_district'))
```

```python
discrepancies = merged_data[
    (merged_data['total_transactions_state'] != merged_data['total_transactions_district']) |
    (merged_data['total_users_state'] != merged_data['total_users_district']) |
    (merged_data['total_amount_state'] != merged_data['total_amount_district'])]
```
